```python
from matplotlib import pyplot as plt
import numpy as np


i_iter = 10
ahc_spinor = np.load(f"results/spinor-ahc_int_iter-{i_iter:04d}.npz")
ahc_soc = np.load(f"results/soc-ahc_int_iter-{i_iter:04d}.npz")

plt.plot(ahc_spinor["Energies_0"], ahc_spinor["data"][:,2], label="spinor")
plt.plot(ahc_soc["Energies_0"], ahc_soc["data"][:,2], label="soc")
plt.legend()
plt.xlabel("E-EF (eV)")
plt.ylabel("AHC (S/m)")
```
